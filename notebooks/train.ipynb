# training yolo11n

```python
%cd ..

from ultralytics import YOLO
import os

# Load a YOLO11n model
model = YOLO("yolo11n.pt")

# train the model 
device = "mps" # change to your device cuda or mps(for mac)
path = os.path.join("potholes-detect-358" ,"data.yaml")
results = model.train(data = path, epochs=40, imgsz=320
                    ,device= device)
```

```python
# testing the model

from PIL import Image

path_image = os.path.join("potholes-detect-358" ,"test" ,"images"
                    ,"7_jpg.rf.3b547a8824468a91124d60f25706a7b4.jpg")
path_model = os.path.join("runs" ,"detect" ,"train" 
                          ,"weights" ,"best.pt")

model = YOLO(path_model)
results = model.predict(path_image)[0]
Image.fromarray(results.plot())
```

```python
%cd ..

from ultralytics import YOLO
import os

# Load a YOLO11n model
model = YOLO("yolo11s.pt")

# train the model
device = "mps" # change to your device cuda or mps(for mac)
path = os.path.join("potholes-detect-358" ,"data.yaml")
results = model.train(data = path, epochs = 20, imgsz=320
                    ,device=device)
```
